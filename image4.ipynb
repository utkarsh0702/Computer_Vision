```python
#Arithematic operation on image
img = cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg')
img1= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
add= img1+150
sub= img1-150
mult= img1*2
div= img1/2
title=['Original','Add','Sub','Mult','Div']
image=[img1, add, sub, mult, div]
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
