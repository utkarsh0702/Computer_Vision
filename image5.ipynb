```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg',1)
img1=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#for negative
img3= abs(255-img1)
img4= abs(255-img2)
title=['Original','Grey','Negative of Original','Negative of Grey']
image=[img1, img2, img3, img4]
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
# scaling an image
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg',1)
output= cv2.resize(img, None, fx=0.5, fy=1.5, interpolation= cv2.INTER_CUBIC)
cv2.imshow('Output', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Otsu thresholding
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg',0)
thres=0
t_value, b1= cv2.threshold(img,thres,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
t_value, b2= cv2.threshold(img,thres,255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
t_value, b3= cv2.threshold(img,thres,255, cv2.THRESH_TOZERO+cv2.THRESH_OTSU)
t_value, b4= cv2.threshold(img,thres,255, cv2.THRESH_TOZERO_INV+cv2.THRESH_OTSU)
title=['Original','Binary','Binary Inverse','Tozero','Tozero Inverse']
image=[img, b1, b2, b3, b4]
for i in range(5):
    plt.subplot(3,2,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
#Adaptive thresholding
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg',0)
block_size=19
constant=2
th1= cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
th2= cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)
title=['Original','Mean Adaptive Threshold','Gaussian Adaptive Threshold ']
image=[img, th1, th2]
for i in range(3):
    plt.subplot(3,2,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
