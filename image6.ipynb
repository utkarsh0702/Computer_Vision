```python
import cv2
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
# Noise in an image
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg')
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
row, column, channel= img1.shape
p=0.05
output= np.zeros(img1.shape, np.uint8)
for i in range(row):
    for j in range(column):
        r= random.random()
        if r< p/2:
            #pepper sprinkle
            output[i][j]= [0,0,0]
        elif r<p:
            #salted sprinkle
            output[i][j]= [255,255,255]
        else:
            output[i][j]= img1[i][j]

title=['Original','Salted and Pepper']
image=[img1, output]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
#Kernal image and covolution
# 1. Edge showcasing
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg')
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
k= np.array(([-1,-1,-1],[-1,8,-1],[-1,-1,-1]), np.float32)
output= cv2.filter2D(img1, -1, k)
title=['Original','Edged']
image=[img1, output]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
# 3. Sharping an image
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg')
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
k= np.array(([0,-1,0],[-1,5,-1],[0,-1,0]), np.float32)
output= cv2.filter2D(img1, -1, k)
title=['Original','Sharpened']
image=[img1, output]
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
# Median Blur of the noise
img= cv2.imread(r'C:\Users\Utkarsh\Desktop\beautiful_scenery_04_hd_pictures_166258.jpg')
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
row, column, channel= img1.shape
p=0.05
output= np.zeros(img1.shape, np.uint8)
for i in range(row):
    for j in range(column):
        r= random.random()
        if r< p/2:
            #pepper sprinkle
            output[i][j]= [0,0,0]
        elif r<p:
            #salted sprinkle
            output[i][j]= [255,255,255]
        else:
            output[i][j]= img1[i][j]

denoise= cv2.medianBlur(output, 5)

title=['Original','Salted and Pepper', 'Median Blur applied on the noise']
image=[img1, output, denoise]
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(image[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
